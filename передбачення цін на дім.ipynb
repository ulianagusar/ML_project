```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
train.fillna(train.mean(), inplace=True)
train.columns = train.columns.astype(str)
# Розбиваємо дані на тренувальні та тестові
X_train, X_test, y_train, y_test = train_test_split( train.drop("SalePrice",axis=1),train["SalePrice"], test_size=0.2, random_state=42)

# Створюємо об'єкт лінійної регресії та навчаємо його на тренувальних даних
reg = LinearRegression()
reg.fit(X_train, y_train)
# Оцінюємо точність моделі на тестових даних
score = reg.score(X_test, y_test)

# Виводимо точність моделі на екран
print("Accuracy:", score)
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()

# відбираємо 5 найбільш інформативних ознак
selector = SelectKBest(f_regression, k=5)

# створюємо поліноміальні ознаки до ступеня 2
poly_features = PolynomialFeatures(degree=2, include_bias=False)

# створюємо модель лінійної регресії з регуляризацією L1
model = Lasso(alpha=0.1, fit_intercept=True)

# створюємо піплайн для об'єднання кроків обробки даних та моделі
pipeline = make_pipeline(scaler, selector, poly_features, model)

# навчаємо модель на тренувальних даних
pipeline.fit(X_train, y_train)

# оцінюємо продуктивність моделі на тестових даних
score = pipeline.score(X_test, y_test)

print("Коефіцієнт детермінації: {:.2f}".format(score))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import make_pipeline

# нормалізуємо вхідні дані
scaler = StandardScaler()

# відбираємо 5 найбільш інформативних ознак
selector = SelectKBest(f_regression, k=5)

# створюємо поліноміальні ознаки до ступеня 2
poly_features = PolynomialFeatures(degree=2, include_bias=False)

# створюємо модель лінійної регресії з регуляризацією L1
model = LinearRegression()

# створюємо піплайн для об'єднання кроків обробки даних та моделі
pipeline = make_pipeline(scaler, selector, poly_features, model)

# навчаємо модель на тренувальних даних
pipeline.fit(X_train, y_train)

# оцінюємо продуктивність моделі на тестових даних
score = pipeline.score(X_test, y_test)

print("Коефіцієнт детермінації: {:.2f}".format(score))
```
