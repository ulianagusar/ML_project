```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()

# Ð²Ñ–Ð´Ð±Ð¸Ñ€Ð°Ñ”Ð¼Ð¾ 5 Ð½Ð°Ð¹Ð±Ñ–Ð»ÑŒÑˆ Ñ–Ð½Ñ„Ð¾Ñ€Ð¼Ð°Ñ‚Ð¸Ð²Ð½Ð¸Ñ… Ð¾Ð·Ð½Ð°Ðº
selector = SelectKBest(f_regression, k=5)

# ÑÑ‚Ð²Ð¾Ñ€ÑŽÑ”Ð¼Ð¾ Ð¿Ð¾Ð»Ñ–Ð½Ð¾Ð¼Ñ–Ð°Ð»ÑŒÐ½Ñ– Ð¾Ð·Ð½Ð°ÐºÐ¸ Ð´Ð¾ ÑÑ‚ÑƒÐ¿ÐµÐ½Ñ 2
poly_features = PolynomialFeatures(degree=2, include_bias=False)

# ÑÑ‚Ð²Ð¾Ñ€ÑŽÑ”Ð¼Ð¾ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð»Ñ–Ð½Ñ–Ð¹Ð½Ð¾Ñ— Ñ€ÐµÐ³Ñ€ÐµÑÑ–Ñ— Ð· Ñ€ÐµÐ³ÑƒÐ»ÑÑ€Ð¸Ð·Ð°Ñ†Ñ–Ñ”ÑŽ L1
model = Lasso(alpha=0.1, fit_intercept=True)

# ÑÑ‚Ð²Ð¾Ñ€ÑŽÑ”Ð¼Ð¾ Ð¿Ñ–Ð¿Ð»Ð°Ð¹Ð½ Ð´Ð»Ñ Ð¾Ð±'Ñ”Ð´Ð½Ð°Ð½Ð½Ñ ÐºÑ€Ð¾ÐºÑ–Ð² Ð¾Ð±Ñ€Ð¾Ð±ÐºÐ¸ Ð´Ð°Ð½Ð¸Ñ… Ñ‚Ð° Ð¼Ð¾Ð´ÐµÐ»Ñ–
pipeline = make_pipeline(scaler, selector, poly_features, model)

# Ð½Ð°Ð²Ñ‡Ð°Ñ”Ð¼Ð¾ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð½Ð° Ñ‚Ñ€ÐµÐ½ÑƒÐ²Ð°Ð»ÑŒÐ½Ð¸Ñ… Ð´Ð°Ð½Ð¸Ñ…
pipeline.fit(X_train, y_train)

# Ð¾Ñ†Ñ–Ð½ÑŽÑ”Ð¼Ð¾ Ð¿Ñ€Ð¾Ð´ÑƒÐºÑ‚Ð¸Ð²Ð½Ñ–ÑÑ‚ÑŒ Ð¼Ð¾Ð´ÐµÐ»Ñ– Ð½Ð° Ñ‚ÐµÑÑ‚Ð¾Ð²Ð¸Ñ… Ð´Ð°Ð½Ð¸Ñ…
score = pipeline.score(X_test, y_test)

print("ÐšÐ¾ÐµÑ„Ñ–Ñ†Ñ–Ñ”Ð½Ñ‚ Ð´ÐµÑ‚ÐµÑ€Ð¼Ñ–Ð½Ð°Ñ†Ñ–Ñ—: {:.2f}".format(score))
```
